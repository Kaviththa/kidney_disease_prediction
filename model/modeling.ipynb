```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head().T
```

```python
df.nunique()
```

```python
columns_to_scaled = ['age','bgr','bu','sc','sod','pot','hemo','pcv','wc']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmaxscaler = MinMaxScaler()
df[columns_to_scaled ] = minmaxscaler .fit_transform(df[columns_to_scaled ])
df.head()
```

```python
df.iloc[0:1,]
```

```python
minmaxscaler.transform([['48','121','36','1.2','137','4','15','44','7800']])
```

```python
df['bp'] = df['bp']/100
```

```python
df.head()
```

```python
X = df.drop('classification',axis=1)
y = df['classification']

#feature selection
from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(X, y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
KN_clf = KNeighborsClassifier(n_neighbors=5)
KN_clf.fit(X_train,y_train)
```

```python
KN_predicted = KN_clf.predict(X_test)
```

```python
KN_predicted
```

```python
KN_clf.score(X_test,y_test)
```

```python
print(confusion_matrix(KN_predicted,y_test))
```

```python

SVM_clf = SVC()
SVM_clf.fit(X_train,y_train)
```

```python
SVM_predicted = SVM_clf.predict(X_test)
```

```python
SVM_clf.score(X_test,y_test)
```

```python
print(confusion_matrix(SVM_predicted,y_test))
```
