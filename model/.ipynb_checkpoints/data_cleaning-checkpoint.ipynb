```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('kidney_disease.csv')
```

```python
df.head()
```

```python
df.head().T
```

```python
df.info()
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
for feature in ['rc','wc','pcv']:
    df[feature] = pd.to_numeric(df[feature],errors='coerce')
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values()
```

```python
((df.isnull().sum()/df.shape[0])*100).sort_values(ascending=False).plot(kind='bar', figsize=(10,10))
```

```python
df.drop(['rbc','rc','pc'],axis=1,inplace=True)
```

```python
df.isnull().sum().sort_values()
```

```python
#extract numerical and categarical 
def extract_cat_num(df):
    cat_col=[col for col in df.columns if df[col].dtype=='object']
    num_col=[col for col in df.columns if df[col].dtype!='object']
    return cat_col,num_col
```

```python
cat_col,num_col= extract_cat_num(df)
```

```python
for col in cat_col:
    print('{} has {} values '.format(col,df[col].unique()))
    print('\n')
```

```python
#Replace incorrect values

df['dm'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

df['cad'] = df['cad'].replace(to_replace = '\tno', value='no')

df['classification'] = df['classification'].replace(to_replace = 'ckd\t', value = 'ckd')

#dataset['classification']=dataset['classification'].replace(to_replace={'ckd':1.0,'ckd\t':1.0,'notckd':0.0,'no':0.0})
```

```python
for col in cat_col:
    print('{} has {} values '.format(col,df[col].unique()))
    print('\n')
```

```python
for col in num_col:
    print('{} has {} values '.format(col,df[col].unique()))
    print('\n')
```

```python
for i in ['age','bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc']:
      df[i].fillna(df[i].mean(),inplace=True)
```

```python
df.isnull().sum().sort_values()
```

```python
df.dropna(subset=['ane','appet','pe','htn','dm','cad','pcc','ba','bp'],axis=0,inplace=True)
```

```python
df.isnull().sum().sort_values()
```

```python
for col in cat_col:
    print('{} has {} values '.format(col,df[col].unique()))
    print('\n')
```

```python
df['classification'] = df['classification'].replace(to_replace={'ckd':1,'notckd':0})
df['pcc']           = df['pcc'].replace(to_replace={'present':1,'notpresent':0})
df['ba']             = df['ba'].replace(to_replace={'present':1,'notpresent':0})

for i in ['htn','dm','cad','pe','ane']:
    df[i]=df[i].replace(to_replace={'yes':1,'no':0})

df['appet']=df['appet'].replace(to_replace={'good':1,'bad':0})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))
for i,feature in enumerate(cat_col):
    plt.subplot(4,3,i+1)
    sns.countplot(df[feature])
```

```python
plt.figure(figsize=(30,20))
for i,feature in enumerate(num_col):
    plt.subplot(5,3,i+1)
    df[feature].hist()
    plt.title(feature)
```
