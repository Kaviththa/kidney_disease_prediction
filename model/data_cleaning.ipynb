```python
#extract numerical and categarical 
def extract_cat_num(df):
    cat_col=[col for col in df.columns if df[col].dtype=='object']
    num_col=[col for col in df.columns if df[col].dtype!='object']
    return cat_col,num_col
```

```python
cat_col,num_col= extract_cat_num(df)
```

```python
for col in cat_col:
    print('{} has {} values '.format(col,df[col].unique()))
    print('\n')
```

```python
#Replace incorrect values

df['dm'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

df['cad'] = df['cad'].replace(to_replace = '\tno', value='no')

df['classification'] = df['classification'].replace(to_replace = 'ckd\t', value = 'ckd')

#dataset['classification']=dataset['classification'].replace(to_replace={'ckd':1.0,'ckd\t':1.0,'notckd':0.0,'no':0.0})
```

```python
for i in ['age','bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc']:
      df[i].fillna(df[i].mean(),inplace=True)
```

```python
df.dropna(subset=['ane','appet','pe','htn','dm','cad','pcc','ba','bp'],axis=0,inplace=True)
```

```python
df['classification'] = df['classification'].replace(to_replace={'ckd':1,'notckd':0})
df['pcc']           = df['pcc'].replace(to_replace={'present':1,'notpresent':0})
df['ba']             = df['ba'].replace(to_replace={'present':1,'notpresent':0})

for i in ['htn','dm','cad','pe','ane']:
    df[i]=df[i].replace(to_replace={'yes':1,'no':0})

df['appet']=df['appet'].replace(to_replace={'good':1,'poor':0})
```

```python
def fill_null_with_predicted_values(df,feature):
   
    
    df_train = df[df[feature].notnull()]
    df_train_X = df_train.drop(feature,axis=1)
    df_train_y = df_train[feature]
    df_test = df[df[feature].isnull()]
    df_test_X = df_test.drop(feature,axis=1)
    
    from sklearn.ensemble import RandomForestClassifier
    feature_clf = RandomForestClassifier(n_estimators=20)
    
    feature_clf.fit(df_train_X,df_train_y)
    df_test_X[feature] = feature_clf.predict(df_test_X)
    
    df_1[feature].fillna(df_test_X[feature],inplace=True)
    
    return df_1.isnull().sum()
     
    
```
